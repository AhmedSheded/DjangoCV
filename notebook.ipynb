```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('dog.png')
```

```python
img.shape
```

```python
normal = cv.cvtColor(img, cv.COLOR_BGR2RGB)
```

```python
def get_image(img, cmap=None):
    plt.axis('off')
    plt.imshow(img, cmap=cmap)
    plt.show()
```

```python
get_image(normal)
```

```python
hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)
get_image(hsv)
```

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
get_image(gray, cmap='gray')
```

```python
# Blurred

width, height = normal.shape[:2]
n = int(0.05 * width)
k = (n, n)
blur = cv.blur(normal, k)
get_image(blur)
```

```python
#binary
v, binary = cv.threshold(gray, 100, 255, cv.THRESH_BINARY)
get_image(binary, cmap='gray')
```

```python
#invert
invert = cv.bitwise_not(binary)
get_image(invert, cmap='gray')
```
